```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
detection = cv2.CascadeClassifier('cascadeTic.xml')
```

```python
def adjusted_detect(img):
     
    detect_img = img.copy()
     
    detect_rect = detection.detectMultiScale(detect_img,
                                              scaleFactor = 1.2,
                                              minNeighbors = 5)
     
    for (x, y, w, h) in detect_rect:
        cv2.rectangle(detect_img, (x, y),
                      (x + w, y + h), (255, 255, 0), 10)\
         
    return detect_img
 
```

```python
img = cv2.imread('test/4_1635679262.1919606.png')
```

```python
produit = adjusted_detect(img)
plt.imshow(produit)
```
